```python
# --- IMPORT TOOLS
# Pandas for working with DataFrames
import pandas as pd
# Import numpy for high-perfomance mathematical and numerical operations
import numpy as np
# Import Matplotlib for creating charts and visualizations
import matplotlib.pyplot as plt
# Import Seaborn for creating attractive statistical visualizations with less code
import seaborn as sns

# This makes plots appear directly inside the notebook
%matplotlib inline

print("✅ Libreries loaded succesfully")
```

```python
# --- LOAD THE DATASET
# pd.read_csv() reads the file and converts it into a DataFrame.
# encoding="latin-1" is required because the file contains special characters.
df = pd.read_csv("../data/raw/sales.csv", encoding="latin-1")

# .shape returns the number of rows and columns in the DataFrame.
print(f"The dataset has {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# --- PREVIEW THE DATASET
# .head(10) displays the first 10 rows of the DataFrame.
# This helps you quickly inspect the structure and contents of a new dataset.
df.head(10)
```

```python
# --- GENERAL INFORMATION
# .info views the "identity file" from the DataFrame
# - how many rows there is in total
# - how many columns exists
# - what kind of data type has every column(int,float,object)
# - how many values are null in every column
df.info()
```

```python
# --- BASIC STATISTICS
# .describe() generates descriptive statistics for all numerical columns,
# including count, mean, standard deviation, minimum, maximum,
# and the 25th, 50th (median), and 75th percentiles.
df.describe()
```

```python
# --- CHECK FOR MISSING VALUES
# .isnull() creates a boolean mask indicating missing values.
# .sum() counts the number of missing values in each column
# because True is treated as 1 and False as 0.
# The result shows the total number of NaN values per column.
df.isnull().sum()
```

```python
# --- CHECK FOR EXACT DUPLICATES
# .duplicated() marks True for each row that is an exact duplicate
# of a previous row.
# .sum() counts the total number of duplicated rows.
duplicates = df.duplicated().sum()

print(f"Duplicate rows found: {duplicates}")
```

```python
# --- CHECK DATA TYPES
# .dtypes displays the data type of each column,
# such as int64, float64, object (text), etc.
# We are especially interested in the date column.
# It is currently stored as an "object" and will be converted
# to a proper datetime type in the next step.
df.dtypes
```

```python
# --- CHECK FOR HIDDEN MISSING VALUES

# Select all text columns (object and string dtypes).
# Including both types avoids compatibility warnings
text_columns = df.select_dtypes(include=["object", "string"]).columns

for col in text_columns:
    # .isin() checks whether a value matches one of the
    # common placeholders used to represent missing data.
    suspicious_values = df[col].isin(   
        ["N/A", "n/a", "-", "", " ", "NULL", "null"]
    ).sum()

    if suspicious_values > 0:
        print(
            f"Column '{col}': {suspicious_values} suspicious values found"
        )

print("Review completed ✅")
```

```python
# --- CONVERT TEXT DATES TO DATETIME

# pd.to_datetime() converts a string such as "2023-11-15"
# into a proper datetime object.
# Once a column has a datetime type, pandas can extract
# components such as year, month, day, and more.
# detect the date format.

df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

# Verify that the conversion was successful.
print(df[["Order Date", "Ship Date"]].dtypes)
```

```python
# --- LOAD THE CLEAN DATASET
# Always work from the processed/ folder, never from raw/
# parse_dates tells pandas to convert these columns to datetime while loading
# This way, we don't need to perform the conversion again

df = pd.read_csv(
    "../data/processed/ventas_limpio.csv",
    parse_dates=["Order Date", "Ship Date"]
)

print(f"Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")
```

```python
# --- SALES TREND BY YEAR
# .groupby() groups rows that share the same value in a column
# It works like SQL's "GROUP BY"
# .agg() defines what to calculate for each group:
#   - Sales: total sales amount
#   - Order ID: count the number of orders
# .reset_index() converts the result into a regular DataFrame
annual_sales = df.groupby("Order Year").agg(
    total_sales=("Sales", "sum"),
    number_of_orders=("Order ID", "count")
).reset_index()

print(annual_sales)
```

```python
# --- PROFIT BY CATEGORY 
# Same pattern: groupby + agg
# This time, we group by Category and calculate the total Profit
# .sort_values() sorts the result from highest to lowest (ascending=False)

profit_by_category = df.groupby("Category").agg(
    Total_Profit=("Profit", "sum"),
    Total_Sales=("Sales", "sum")
).reset_index().sort_values("Total_Profit", ascending=False)

print(profit_by_category)
```

```python
# --- SALES BY REGION
# Group the data by Region
# Calculate the total Sales and total Profit for each region
# Sort the results from highest to lowest total sales

sales_by_region = df.groupby("Region").agg(
    Total_Sales=("Sales", "sum"),
    Total_Profit=("Profit", "sum")
).reset_index().sort_values("Total_Sales", ascending=False)

print(sales_by_region)
```

```python
# --- DISCOUNT VS PROFIT
# pd.cut() splits the Discount values into ranges (bins)
# For example: [0–20%], [20–40%], [40–60%], [60–80%]
# This allows us to compare discount groups with their average profit

df["Discount Range"] = pd.cut(
    df["Discount"],
    bins=[0, 0.2, 0.4, 0.6, 0.8],  # The boundaries of each range
    labels=["0-20%", "20-40%", "40-60%", "60-80%"],  # Labels for each range
    include_lowest=True  # Include the value 0 in the first range
)

discount_vs_profit = df.groupby("Discount Range", observed=True).agg(
    Average_Profit=("Profit", "mean"),
    Number_of_Transactions=("Order ID", "count")
).reset_index()

print(discount_vs_profit)
```

```python
# ── Celda 19: Configuración visual global ────────────────────────────
# seaborn tiene "temas" predefinidos que mejoran la estética automáticamente
# "whitegrid" pone fondo blanco con líneas de cuadrícula grises — ideal para datos
sns.set_theme(style="whitegrid")

# Definimos un tamaño estándar para todos los gráficos
# figsize=(ancho, alto) en pulgadas
plt.rcParams["figure.figsize"] = (10, 5)

print("✅ Configuración visual lista")
```

```python
# ── Celda 20: Gráfico 1 — Ventas anuales (barras) ────────────────────
# plt.figure() crea un lienzo nuevo para el gráfico
fig, ax = plt.subplots(figsize=(8, 5))

# sns.barplot dibuja barras — x es la categoría, y es el valor numérico
# palette define los colores — "Blues_d" es una paleta de azules
sns.barplot(
    data=annual_sales,
    x="Order Year",
    y="total_sales",
    hue="Order Year",  # agrega esto
    palette="Blues_d",
    legend=False,      # y esto
    ax=ax
)

# Etiquetas y título — sin esto el gráfico no comunica nada
ax.set_title("Ventas Totales por Año", fontsize=14, fontweight="bold")
ax.set_xlabel("Año")
ax.set_ylabel("Ventas (USD)")

# Formato: convierte 200000 en "$200,000" para que sea legible
ax.yaxis.set_major_formatter(
    plt.FuncFormatter(lambda x, _: f"${x:,.0f}")
)

plt.tight_layout()  # evita que las etiquetas se corten
plt.savefig("../outputs/graphics/annual_sales.png", dpi=150)  # guarda la imagen
plt.show()
```

```python
# ── Celda 21: Gráfico 2 — Ganancia por categoría ─────────────────────
fig, ax = plt.subplots(figsize=(8, 5))

# Los colores comunican: verde = positivo, rojo = negativo o bajo
colores = ["#2ecc71", "#3498db", "#e74c3c"]

sns.barplot(
    data=profit_by_category,
    x="Category",
    y="Total_Sales",
    hue = "Category",
    palette=colores,
    ax=ax
)

ax.set_title("Ganancia Total por Categoría", fontsize=14, fontweight="bold")
ax.set_xlabel("Categoría")
ax.set_ylabel("Ganancia (USD)")
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x:,.0f}"))

# Línea horizontal en 0 para que sea fácil ver si algo está en negativo
ax.axhline(0, color="black", linewidth=0.8, linestyle="--")

plt.tight_layout()
plt.savefig("../outputs/graphics/ganancia_categoria.png", dpi=150)
plt.show()
```

```python
# ── Celda 22: Gráfico 3 — Descuento vs Ganancia promedio ─────────────
# Este es el gráfico más importante del análisis — muestra el hallazgo clave
fig, ax = plt.subplots(figsize=(8, 5))

# Usamos colores que refuercen el mensaje: rojo en los rangos donde se pierde dinero
colores_descuento = ["#2ecc71", "#f39c12", "#e67e22", "#e74c3c"]

sns.barplot(
    data=discount_vs_profit,
    x="Discount Range",
    y="Average_Profit",
    hue="Discount Range",
    palette=colores_descuento,
    ax=ax
)

ax.set_title("Ganancia Promedio por Rango de Descuento", fontsize=14, fontweight="bold")
ax.set_xlabel("Rango de Descuento")
ax.set_ylabel("Ganancia Promedio (USD)")
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x:,.0f}"))
ax.axhline(0, color="black", linewidth=0.8, linestyle="--")

plt.tight_layout()
plt.savefig("../outputs/graphics/descuento_vs_ganancia.png", dpi=150)
plt.show()
```

```python
# ── Celda 23: Gráfico 4 — Ventas por región ──────────────────────────
fig, ax = plt.subplots(figsize=(8, 5))

sns.barplot(
    data=sales_by_region.sort_values("Total_Sales", ascending=False),
    x="Region",
    y="Total_Sales",
    hue = "Region",
    palette="Greens_d",
    ax=ax
)

ax.set_title("Ventas Totales por Región", fontsize=14, fontweight="bold")
ax.set_xlabel("Región")
ax.set_ylabel("Ventas (USD)")
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x:,.0f}"))

plt.tight_layout()
plt.savefig("../outputs/graphics/ventas_region.png", dpi=150)
plt.show()
```
